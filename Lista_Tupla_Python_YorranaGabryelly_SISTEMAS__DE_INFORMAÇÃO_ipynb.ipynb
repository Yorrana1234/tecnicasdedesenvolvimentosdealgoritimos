```python
alunos_db = [
  {"matricula": "2025001", "nome": "João Lucas", "curso": "Medicina"},
  {"matricula": "2025002", "nome": "Ana Julia", "curso": "Sistemas De Informação"},
  {"matricula": "2025003", "nome": "Lucas Silva", "curso": "Enfermagem"},
]
print("1. Sistemas Simplificado De Cadastro De Alunos")
print("-" * 40)

# 1. Adicionar Novo Aluno (CREATE)
print("-> Adicionando novo aluno...")
novo_aluno = {"matricula": "2025004", "nome": "Ana Clara", "curso": "Engenharia De Software"}
alunos_db.append(novo_aluno) #Adiciona o dicionario (novo_aluno) a lista alunos_db
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso!")

# 2. Buscar Aluno por Matricula (READ)
matricula_busca = "2025005"
aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)

if aluno_encontrado:
  print(f"\nAluno encontrado (matricula: {matricula_busca}):")
  # Acessa os valores do dicionario usando suas chaves.
  print(f"Nome: {aluno_encontrado['nome']}, Curso: {aluno_encontrado['curso']}")
else:
  print(f"\nAluno com matricula {matricula_busca} não encontrado.")

# 3. Atualizar Informações do Curso (UPDATE)
matricula_update = "2025003" # Changed to a matricula that exists in the initial list
for aluno in alunos_db:
  if aluno["matricula"] == matricula_update:
    aluno["curso"] = "Segurança da Informação"
    print(f"\nCurso do aluno {aluno['nome']} atualizado para: {aluno['curso']}")
    break
print("-" * 40)

# Display the updated list of students (optional)
print("\nLista de alunos atualizada:")
for aluno in alunos_db:
    print(aluno)
```

```python
"TAREFA 04"
"Faça um programa que solicita uma quantidade indefinida de notas de uma prova (deve parar quando um númeronegativo for digitado). Armazena em uma lista. Após isso, o programa deve mostrar a média das notas."

notas = []
while True:
  nota = float(input("Digite uma nota (ou número negativo para sair): "))
  if nota < 0:
    break
  notas.append(nota)

if len(notas) > 0:
    media = sum(notas) / len(notas)
    print(f"\nNotas digitadas: {notas}")
    print(f"Média das notas: {media:.2f}")
else:
    print("Nenhuma nota válida foi digitada.")
```

```python
"TAREFA 05"
"Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado). Armazene-os em duas listas:  uma para os números pares; outra para os números ímpares. A seguir, mostre a porcentagem de pares e ímpares digitados"

while True:
  num = int(input("Digite um número positivo (ou número negativo para encerrar): "))
  if num <0:
    break
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)
        total = len(pares) + len(impares)
        print("\n--- Resultados ---")
if total > 0:
    perc_pares = (len(pares) / total) * 100
    perc_impares = (len(impares) / total) * 100

    print(f"Números pares: {pares}")
    print(f"Números ímpares: {impares}")
    print(f"Porcentagem de pares: {perc_pares:.2f}%")
    print(f"Porcentagem de ímpares: {perc_impares:.2f}%")
else:
    print("Nenhum número válido foi digitado.")
```

```python
"TAREFA 06"
"Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado). Armazene todos os números digitados em uma lista, sem repetição. Exercícios"
numeros = []
while True:
    try:
        num_str = input("Digite um número positivo (ou número negativo para encerrar): ")
        num = int(num_str)
        if num < 0:
            break
        if num >= 0 and num not in numeros:
            numeros.append(num)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

print("\n--- Resultados ---")
if len(numeros) > 0:
    print(f"Números digitados (sem repetição): {numeros}")
else:
    print("Nenhum número válido foi digitado.")
```

```python
"TAREFA 07"
"Escreva um programa que solicita que o usuário digite a altura e o peso de 5 pessoas. Verifique se pelo menos duas das pessoas tem a mesma altura e mesmo peso"

pessoas = [] # Initialize the list 'pessoas' here
for i in range(1, 6):
    print(f"\nPessoa {i}:")
    altura = float(input("Digite a altura (em metros): "))
    peso = float(input("Digite o peso (em kg): "))
    pessoas.append((altura, peso)) # Append the tuple to the list

duplicado = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i] == pessoas[j]:
            duplicado = True
            break
    if duplicado:
        break

print("\n--- Resultado ---")
if duplicado:
    print("Há pelo menos duas pessoas com a mesma altura e o mesmo peso.")
else:
    print("Nenhuma combinação de altura e peso se repete entre as pessoas.")
```
